```python
import pandas as pd
import os
import glob

# ১. ল্যাব ডেটাকে পিভট করা (এটি একটু সময় নিতে পারে)
print("Pivoting lab results... Making 65 parameter columns.")
labs_pivoted = filtered_labs.pivot_table(
    index=['SUBJECT_ID', 'HADM_ID'], 
    columns='ITEMID', 
    values='VALUENUM', 
    aggfunc='mean'
).reset_index()

# ২. ডায়াগনোসিস মাস্টার ফাইলের সাথে মার্জ করা
final_master = pd.merge(master_v3, labs_pivoted, on=['SUBJECT_ID', 'HADM_ID'], how='left')

# ৩. মিসিং ৪টি রোগ (Dengue, Typhoid, Chikungunya, Cholera) রিকভারি
missing_targets = ['Dengue', 'Typhoid', 'Chikungunya', 'Cholera']
recovery_list = []

print("Searching for missing tropical diseases in all files...")
all_csvs = glob.glob('data/*.csv') + glob.glob('*.csv')

for file in all_csvs:
    # আমাদের মাস্টার ফাইলগুলো বাদ দিচ্ছি সার্চ থেকে
    if 'SYNAPSE_MASTER' in file or 'SYNAPSE_FINAL' in file: continue
    
    try:
        temp_df = pd.read_csv(file, nrows=500) # দ্রুত চেক করার জন্য
        # রোগের কলাম খুঁজে বের করা (যেখানে 'diagnosis' বা 'disease' শব্দ আছে)
        disease_col = [c for c in temp_df.columns if any(k in c.lower() for k in ['disease', 'diagnosis', 'type'])]
        
        if disease_col:
            full_df = pd.read_csv(file)
            col_name = disease_col[0]
            # মিসিং রোগগুলো এই ফাইলে আছে কি না চেক করা
            found_df = full_df[full_df[col_name].str.contains('|'.join(missing_targets), case=False, na=False)].copy()
            if not found_df.empty:
                found_df['Disease_Type'] = found_df[col_name]
                recovery_list.append(found_df)
                print(f"Found missing diseases in: {file}")
    except:
        continue

# ৪. সব ডেটা একসাথে করা
if recovery_list:
    recovered_data = pd.concat(recovery_list, ignore_index=True)
    final_master = pd.concat([final_master, recovered_data], ignore_index=True)

# ৫. ডুপ্লিকেট পরিষ্কার করা এবং সেভ করা
final_master = final_master.drop_duplicates(subset=['SUBJECT_ID', 'Disease_Type'])
print(f"\n--- SYNAPSE OS: FINAL STATUS ---")
print(f"Total Unique Diseases: {final_master['Disease_Type'].nunique()} / 44")
print(f"Total Rows: {len(final_master)}")

final_master.to_csv('data/SYNAPSE_FINAL_MASTER_V3.csv', index=False)
```

```python
import pandas as pd
import os
import glob

# ১. মাস্টার ফাইল লোড করা
master_path = 'data/SYNAPSE_FINAL_MASTER_V3.csv'
master_df = pd.read_csv(master_path, low_memory=False)

# ২. আর্কিভ এবং ডাটা ফোল্ডারে সার্চ করা
search_folders = ['data/', 'archive_old_files/']
missing_targets = ['Dengue', 'Typhoid', 'Chikungunya', 'Cholera', 'Malaria']
rescued_dfs = []

print("--- SYNAPSE OS: KAGGLE DATA RECOVERY MISSION ---")

for folder in search_folders:
    if os.path.exists(folder):
        # সব CSV ফাইল চেক করা হচ্ছে
        for file in glob.glob(folder + "*.csv"):
            try:
                # ফাইলটি পড়ে রোগের নাম খোঁজা
                temp_df = pd.read_csv(file, nrows=100)
                disease_col = [c for c in temp_df.columns if any(k in c.lower() for k in ['disease', 'diagnosis', 'type'])]
                
                if disease_col:
                    full_df = pd.read_csv(file)
                    col = disease_col[0]
                    # মিসিং রোগগুলো থাকলে ফিল্টার করা
                    found = full_df[full_df[col].str.contains('|'.join(missing_targets), case=False, na=False)].copy()
                    
                    if not found.empty:
                        found['Disease_Type'] = found[col]
                        # অপ্রয়োজনীয় কলাম বাদ দিয়ে মাস্টার ফরম্যাটে আনা
                        rescued_dfs.append(found)
                        print(f"Rescued {len(found)} records from: {file}")
            except:
                continue

# ৩. উদ্ধার করা ডেটা মাস্টার ডাটাসেটে যোগ করা
if rescued_dfs:
    all_rescued = pd.concat(rescued_dfs, ignore_index=True)
    # নামগুলো স্ট্যান্ডার্ড করা
    all_rescued['Disease_Type'] = all_rescued['Disease_Type'].replace({
        'Dengue Fever': 'Dengue',
        'Typhoid Fever': 'Typhoid'
    })
    
    # বর্তমান মাস্টারের সাথে জোড়া লাগানো
    master_df = pd.concat([master_df, all_rescued], ignore_index=True)

# ৪. ডুপ্লিকেট পরিষ্কার এবং ফাইনাল রিপোর্ট
master_df = master_df.drop_duplicates(subset=['SUBJECT_ID', 'Disease_Type'])
print(f"\n--- MISSION ACCOMPLISHED ---")
print(f"Total Unique Diseases: {master_df['Disease_Type'].nunique()} / 44")
print(f"Final Row Count: {len(master_df)}")

# ফাইনাল মাস্টার সেভ
master_df.to_csv('data/SYNAPSE_FINAL_MASTER_V4.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# ১. ফাইলগুলো আবার লোড করা
print("Reloading files for structural fix...")
master_v5 = pd.read_csv('data/SYNAPSE_FINAL_MASTER_V5.csv', low_memory=False)
labs_v3 = pd.read_csv('data/SYNAPSE_LABS_V3.csv', low_memory=False)

# ২. আইডি-গুলোর টাইপ একই করা (যাতে মার্জিং এ সমস্যা না হয়)
master_v5['SUBJECT_ID'] = master_v5['SUBJECT_ID'].astype(str)
master_v5['HADM_ID'] = master_v5['HADM_ID'].fillna(0).astype(int).astype(str)

labs_v3['SUBJECT_ID'] = labs_v3['SUBJECT_ID'].astype(str)
labs_v3['HADM_ID'] = labs_v3['HADM_ID'].fillna(0).astype(int).astype(str)

# ৩. ল্যাব ডেটাকে পিভট করা (সরাসরি ITEMID ব্যবহার করে)
print("Pivoting lab data correctly...")
labs_pivot = labs_v3.pivot_table(
    index=['SUBJECT_ID', 'HADM_ID'], 
    columns='ITEMID', 
    values='VALUENUM', 
    aggfunc='mean'
).reset_index()

# ৪. মার্জ করা
final_df = pd.merge(master_v5, labs_pivot, on=['SUBJECT_ID', 'HADM_ID'], how='inner')

# ৫. রোগের নামের মিসম্যাচ ফিক্স করা (Fuzzy Matching)
# যদি ফাইলে 'Dengue' থাকে আর আমরা 'Dengue Fever' খুঁজি, তবে সেটা ফিক্স হবে
def clean_disease_names(name):
    name = str(name).lower()
    if 'dengue' in name: return 'Dengue Fever'
    if 'typhoid' in name: return 'Typhoid Fever'
    if 'chikun' in name: return 'Chikungunya'
    if 'cholera' in name: return 'Cholera'
    return name.title()

final_df['Disease_Type'] = final_df['Disease_Type'].apply(clean_disease_names)

# ৬. রিপোর্ট চেক
print("\n--- SYNAPSE OS: RECOVERY REPORT ---")
print(f"Total Rows after Correct Merge: {len(final_df)}")
print(f"Unique Diseases Found: {final_df['Disease_Type'].nunique()}")
print(f"Missing Values in Potassium (ITEMID 50971): {final_df[50971].isnull().sum() if 50971 in final_df.columns else 'Not Found'}")

final_df.to_csv('data/SYNAPSE_FINAL_MASTER_V6.csv', index=False)
```

```python
import pandas as pd
import os

print("--- SYNAPSE OS: THE MEGA DATA EXTRACTION ---")

# ১. ITEMID গুলো আবার ডায়নামিকালি জেনারেট করা (আপডেটেড কি-ওয়ার্ড সহ)
dict_path = 'data/D_LABITEMS.csv.gz' if os.path.exists('data/D_LABITEMS.csv.gz') else 'data/D_LABITEMS.csv'
d_items = pd.read_csv(dict_path, low_memory=False)

target_labs = {
    'Potassium': ['potassium'], 'Sodium': ['sodium'], 'Calcium': ['calcium'],
    'Chloride': ['chloride'], 'BUN': ['bun', 'urea nitrogen'], 'Glucose': ['glucose'],
    'Hemoglobin': ['hemoglobin'], 'WBC': ['white blood cells', 'wbc', 'leukocyte'],
    'PT': ['prothrombin', 'inr', ' pt'], 'APTT': ['aptt', 'thromboplastin'],
    'Bilirubin': ['bilirubin'], 'Magnesium': ['magnesium'], 'Phosphate': ['phosphate'],
    'Creatinine': ['creatinine'], 'Cholesterol': ['cholesterol', 'ldl', 'hdl', 'triglyceride'],
    'Iron': ['tibc', 'iron', 'ferritin']
}

found_itemids = []
for keywords in target_labs.values():
    mask = d_items['LABEL'].astype(str).str.lower().str.contains('|'.join(keywords), na=False)
    found_itemids.extend(d_items[mask]['ITEMID'].tolist())

unique_ids = list(set(found_itemids))
print(f"Loaded {len(unique_ids)} unique ITEMIDs for extraction.")

# ২. LABEVENTS থেকে ডেটা টানা (মেমোরি সেভ করার জন্য শুধু প্রয়োজনীয় কলাম)
print("\nScanning 1.8 Crore rows in LABEVENTS.csv.gz... (This might take 1-2 minutes)")
lab_path = 'data/LABEVENTS.csv.gz'
labs_raw = pd.read_csv(lab_path, compression='gzip', usecols=['SUBJECT_ID', 'ITEMID', 'VALUENUM'])

# শুধুমাত্র আমাদের ১১৪টি আইডির ডেটা ফিল্টার করা
filtered_labs = labs_raw[labs_raw['ITEMID'].isin(unique_ids)].copy()
print(f"Extracted {len(filtered_labs)} relevant lab test records!")

# ৩. 피ভট করা (HADM_ID বাদ দিয়ে শুধু SUBJECT_ID দিয়ে, যাতে কোনো রোগী বাদ না পড়ে)
print("\nPivoting data (converting rows to columns)...")
labs_pivot = filtered_labs.groupby(['SUBJECT_ID', 'ITEMID'])['VALUENUM'].mean().unstack().reset_index()

# ৪. মাস্টার ফাইলের সাথে মার্জ করা
print("\nMerging with the Master Diagnostic File...")
master_v5 = pd.read_csv('data/SYNAPSE_FINAL_MASTER_V5.csv', low_memory=False)

# আইডি টাইপ ঠিক করা
master_v5['SUBJECT_ID'] = pd.to_numeric(master_v5['SUBJECT_ID'], errors='coerce')
labs_pivot['SUBJECT_ID'] = pd.to_numeric(labs_pivot['SUBJECT_ID'], errors='coerce')

# লেফট জয়েন (Left Join) যাতে আমাদের ৪২টি রোগের কোনো রোগী হারিয়ে না যায়
final_mega_df = pd.merge(master_v5, labs_pivot, on='SUBJECT_ID', how='left')

# ৫. ফাইনাল রিপোর্ট
lab_columns = [c for c in final_mega_df.columns if str(c).isdigit()]
print("\n--- SYNAPSE OS: MEGA RECOVERY REPORT ---")
print(f"Total Patient Records: {len(final_mega_df)}")
print(f"Total Lab Parameters (Columns): {len(lab_columns)}")

# ফাঁকা ঘরের পার্সেন্টেজ চেক (আগে যেটা ১০০% ছিল)
if len(lab_columns) > 0:
    sample_col = lab_columns[0]
    missing_pct = final_mega_df[sample_col].isnull().sum() / len(final_mega_df) * 100
    print(f"Missing Value % for ID {sample_col}: {missing_pct:.2f}% (আগে এটা ১০০% ছিল!)")

# ৬. সেভ করা (V7 - The Ultimate Edition)
final_mega_df.to_csv('data/SYNAPSE_FINAL_MASTER_V7.csv', index=False)
print("\nSuccess! Ultimate dataset saved as: data/SYNAPSE_FINAL_MASTER_V7.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report, accuracy_score
import joblib

print("--- SYNAPSE GLOBAL AI: OPTIMIZED TRAINING (2% THRESHOLD) ---\n")

# ১. ডেটা লোড
df = pd.read_csv('data/SYNAPSE_FINAL_MASTER_V7.csv', low_memory=False)

# ২. রোগ ফিল্টারিং (Local tropical diseases saved for Model 2)
disease_counts = df['Disease_Type'].value_counts()
valid_diseases = disease_counts[disease_counts >= 10].index
df_filtered = df[df['Disease_Type'].isin(valid_diseases)].copy()

print(f"Target Diseases for Global AI: {len(valid_diseases)} (Rare/Local diseases kept for Model 2)")

# ৩. প্যারামিটার ফিল্টারিং (জুবায়েরের ২% লজিক!)
X_raw = df_filtered[[col for col in df_filtered.columns if str(col).isdigit()]]
y = df_filtered['Disease_Type']

# ২% এর হিসাব (0.02)
threshold_count = int(0.02 * len(df_filtered)) 
X_kept = X_raw.dropna(axis=1, thresh=threshold_count)
dropped_cols = X_raw.shape[1] - X_kept.shape[1]

print(f"\nTotal Parameters Before: {X_raw.shape[1]}")
print(f"🗑️ Dropped 'Noise' Parameters (< 2% data): {dropped_cols}")
print(f"⭐ Final Golden Parameters Kept: {X_kept.shape[1]}\n")

# ৪. Imputation (ফাঁকা ঘরে Median বসানো)
print("Imputing missing data for Golden Parameters...")
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X_kept)

# ৫. লেবেল এনকোডিং
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# ৬. SMOTE (ডেটা ব্যালেন্স করা)
print("Balancing dataset with SMOTE...")
smote = SMOTE(k_neighbors=5, random_state=42)
X_balanced, y_balanced = smote.fit_resample(X_imputed, y_encoded)

# ৭. ট্রেইন ও টেস্ট স্প্লিট
X_train, X_test, y_train, y_test = train_test_split(
    X_balanced, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced
)

# ৮. XGBoost মডেল ট্রেনিং
print("Training Synapse Global Brain... (M4 is crunching the numbers!)")
model = XGBClassifier(
    n_estimators=150, 
    learning_rate=0.1, 
    max_depth=7, 
    objective='multi:softprob', 
    tree_method='hist',
    random_state=42, 
    n_jobs=-1
)
model.fit(X_train, y_train)

# ৯. ইভ্যালুয়েশন
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("\n" + "="*50)
print(f" 🚀 SYNAPSE GLOBAL MODEL ACCURACY: {accuracy * 100:.2f}% 🚀 ")
print("="*50)

# ১০. রাউটার সিস্টেমের জন্য সেভ করা
joblib.dump(model, 'synapse_global_model.pkl')
joblib.dump(le, 'global_disease_encoder.pkl')
joblib.dump(imputer, 'global_imputer.pkl')
joblib.dump(X_kept.columns.tolist(), 'global_features.pkl')

print("\nSuccess! Synapse Global is optimized and ready.")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_sample_weight
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib

print("--- SYNAPSE OS: RESPIRATORY EXPERT AI (MODEL 5/8) ---\n")

# ১. ডেটা লোড
df = pd.read_csv('data/SYNAPSE_FINAL_MASTER_V7.csv', low_memory=False)

# ২. পিওর ল্যাব-ড্রিভেন ফুসফুসের রোগ (ABG এবং WBC বেসড)
resp_diseases = ['Respiratory Failure', 'Pneumonia']
df_resp = df[df['Disease_Type'].isin(resp_diseases)].copy()

print("Target Diseases for Respiratory AI:")
print(df_resp['Disease_Type'].value_counts())
print(f"\nTotal Respiratory Patients Found: {len(df_resp)}\n")

# ৩. প্যারামিটার ফিল্টারিং 
X_final = df_resp[[col for col in df_resp.columns if str(col).isdigit()]]
y = df_resp['Disease_Type']

print(f"Total ABG & Lab Parameters given to AI: {X_final.shape[1]}")

# ৪. লেবেল এনকোডিং
le_resp = LabelEncoder()
y_encoded = le_resp.fit_transform(y)

# ৫. স্প্লিট
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# অটোমেটিক ওয়েট ক্যালকুলেশন (Binary Classification এর জন্য)
class_counts = np.bincount(y_train)
weight = class_counts[0] / class_counts[1] 

# ৬. মডেল ট্রেনিং
print("Training Respiratory Brain... (Focusing on ABG and Oxygen levels!)")
resp_model = XGBClassifier(
    n_estimators=400,          
    learning_rate=0.05,        
    max_depth=6,              
    objective='binary:logistic', 
    tree_method='hist',
    scale_pos_weight=weight,
    random_state=42, 
    n_jobs=-1,
    missing=np.nan
)

resp_model.fit(X_train, y_train)

# ৭. ইভ্যালুয়েশন
y_pred = resp_model.predict(X_test)
acc_top1 = accuracy_score(y_test, y_pred)

print("\n" + "="*50)
print(f" 🚀 TRUE STRICT ACCURACY : {acc_top1 * 100:.2f}% 🚀")
print("="*50)

print("\nDetailed Clinical Report:")
print(classification_report(y_test, y_pred, target_names=le_resp.classes_))

# ৮. মডেল সেভ করা
joblib.dump(resp_model, 'expert_resp_model.pkl')
joblib.dump(le_resp, 'expert_resp_encoder.pkl')
joblib.dump(X_final.columns.tolist(), 'expert_resp_features.pkl')

print("\nSuccess! Medically accurate Respiratory AI is ready.")
```

```python
import pandas as pd
import numpy as np
import os
import glob
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_sample_weight
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib

print("--- SYNAPSE OS: TROPICAL DATA MERGER & TRAINING (FINAL BOSS) ---\n")

# ১. File Merge Kora
folder_path = 'data' # Tomar data folder er path
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

print(f"🔍 Searching for CSV files in '{folder_path}'...")

df_list = []
for file in all_files:
    try:
        df = pd.read_csv(file, low_memory=False)
        file_name = os.path.basename(file).lower()
        
        # Dengue dataset er jnno target column fix kora
        if 'dengue' in file_name and 'Result' in df.columns:
            df = df.rename(columns={'Result': 'Disease_Type'})
            df['Disease_Type'] = df['Disease_Type'].replace({'positive': 'Dengue', 'negative': 'Normal'})
            df_list.append(df)
            print(f"✅ Loaded Dengue Data: {os.path.basename(file)}")
            
        # Typhoid dataset er jnno target column fix kora
        elif 'typhoid' in file_name or 'processed' in file_name:
            if 'Typhoid' in df.columns:
                df = df.rename(columns={'Typhoid': 'Disease_Type'})
                df['Disease_Type'] = df['Disease_Type'].replace({'Positive': 'Typhoid', 'Negative': 'Normal', 'Minimal': 'Normal'})
                df_list.append(df)
                print(f"✅ Loaded Typhoid Data: {os.path.basename(file)}")
                
        # Malaria ba onnanyo dataset (Synthetic etc.)
        elif 'synthetic' in file_name and 'Malaria_Parasite_Test' in df.columns:
            df['Disease_Type'] = np.where(df['Malaria_Parasite_Test'] == 'Positive', 'Malaria', 'Normal')
            df_list.append(df)
            print(f"✅ Loaded Malaria Data: {os.path.basename(file)}")

    except Exception as e:
        print(f"❌ Error loading {os.path.basename(file)}: {e}")

if not df_list:
    print("\n❌ Kono valid dataset paowa jayni. File gulo ki thikmoto 'data' folder e ache?")
    exit()

# Master dataset toiri kora
df_master = pd.concat(df_list, ignore_index=True)
print(f"\n📊 Total Patients in Tropical Master Dataset: {len(df_master)}")
print("Target Diseases Distribution:")
print(df_master['Disease_Type'].value_counts())

# ২. Model Training er jonno data prostut kora
print("\n⚙️ Preparing data for AI...")
X_final = df_master[[col for col in df_master.columns if str(col).isdigit() or df_master[col].dtype in ['float64', 'int64']]]

if 'Disease_Type' in X_final.columns:
    X_final = X_final.drop(columns=['Disease_Type'])

X_final = X_final.dropna(axis=1, how='all')
y = df_master['Disease_Type'].astype(str)

# Label Encoding
le_trop = LabelEncoder()
y_encoded = le_trop.fit_transform(y)

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

weights = compute_sample_weight(class_weight='balanced', y=y_train)

# ৩. Final Boss Training
print("\n🚀 Training Final Tropical Expert Brain...")
trop_model = XGBClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=7,
    objective='multi:softprob' if len(le_trop.classes_) > 2 else 'binary:logistic',
    tree_method='hist',
    random_state=42,
    n_jobs=-1,
    missing=np.nan
)

trop_model.fit(X_train, y_train, sample_weight=weights)

# ৪. Evaluation
y_pred = trop_model.predict(X_test)
acc_top1 = accuracy_score(y_test, y_pred)

print("\n" + "="*50)
print(f" 🎯 TRUE STRICT ACCURACY : {acc_top1 * 100:.2f}%")
print("="*50)

print("\nDetailed Clinical Report:")
print(classification_report(y_test, y_pred, target_names=le_trop.classes_))

# ৫. Model Save kora
joblib.dump(trop_model, 'expert_tropical_model.pkl')
joblib.dump(le_trop, 'expert_tropical_encoder.pkl')
joblib.dump(X_final.columns.tolist(), 'expert_tropical_features.pkl')

print("\n✅ Success! Tropical AI is ready. The Synapse OS 8-Model Architecture is COMPLETE! 🎉")
```

```python
import pandas as pd
import numpy as np
import os
import glob
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

print("--- SYNAPSE OS: ALL-IN-ONE TROPICAL TRIAGE ---\n")

# 1. Master Data Load & Merge (Directly from 'data' folder)
folder_path = 'data' 
all_files = glob.glob(os.path.join(folder_path, "*.csv"))

print("🔍 Loading and merging raw datasets...")
df_list = []
for file in all_files:
    try:
        df = pd.read_csv(file, low_memory=False)
        file_name = os.path.basename(file).lower()
        
        if 'dengue' in file_name and 'Result' in df.columns:
            df = df.rename(columns={'Result': 'Disease_Type'})
            df['Disease_Type'] = df['Disease_Type'].replace({'positive': 'Dengue', 'negative': 'Normal'})
            df_list.append(df)
        elif 'typhoid' in file_name or 'processed' in file_name:
            if 'Typhoid' in df.columns:
                df = df.rename(columns={'Typhoid': 'Disease_Type'})
                df['Disease_Type'] = df['Disease_Type'].replace({'Positive': 'Typhoid', 'Negative': 'Normal', 'Minimal': 'Normal'})
                df_list.append(df)
        elif 'synthetic' in file_name and 'Malaria_Parasite_Test' in df.columns:
            df['Disease_Type'] = np.where(df['Malaria_Parasite_Test'] == 'Positive', 'Malaria', 'Normal')
            df_list.append(df)
    except:
        pass

if not df_list:
    print("❌ No CSV files found in 'data' folder!")
    exit()

df_master = pd.concat(df_list, ignore_index=True)

# 2. Fix Typhoid String Data (Converting "1:80" to 80)
print("⚙️ Fixing Widal Test Formats (Converting '1:80' to numeric values)...")
for col in df_master.columns:
    if df_master[col].dtype == 'object' and col != 'Disease_Type':
        # "1:80" থেকে শুধু 80 বের করে আনা
        df_master[col] = df_master[col].astype(str).str.extract(r':(\d+)').astype(float)

# ---------------------------------------------------------
# AI 1: DENGUE & MALARIA EXPERT (Blood/CBC Based)
# ---------------------------------------------------------
print("\n🩸 TRAINING AI-1: DENGUE & MALARIA EXPERT...")
df_blood = df_master[df_master['Disease_Type'].isin(['Dengue', 'Malaria', 'Normal'])].copy()
X_blood = df_blood.select_dtypes(include=[np.number]).dropna(axis=1, how='all')
y_blood = LabelEncoder().fit_transform(df_blood['Disease_Type'])

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_blood, y_blood, test_size=0.2, random_state=42, stratify=y_blood)

model_blood = XGBClassifier(n_estimators=300, max_depth=6, tree_method='hist', random_state=42)
model_blood.fit(X_train_b, y_train_b)

acc_blood = accuracy_score(y_test_b, model_blood.predict(X_test_b))
print(f"✅ DENGUE & MALARIA ACCURACY : {acc_blood * 100:.2f}%")

# ---------------------------------------------------------
# AI 2: TYPHOID EXPERT (Widal Test Based)
# ---------------------------------------------------------
print("\n🦠 TRAINING AI-2: TYPHOID EXPERT...")
df_typhoid = df_master[df_master['Disease_Type'].isin(['Typhoid', 'Normal'])].copy()
X_typhoid = df_typhoid.select_dtypes(include=[np.number]).dropna(axis=1, how='all')
y_typhoid = LabelEncoder().fit_transform(df_typhoid['Disease_Type'])

X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_typhoid, y_typhoid, test_size=0.2, random_state=42, stratify=y_typhoid)

# Auto-balancing weights for small Typhoid dataset
class_weights = len(y_train_t) / (len(np.unique(y_train_t)) * np.bincount(y_train_t))
sample_weights = np.array([class_weights[y] for y in y_train_t])

model_typhoid = XGBClassifier(n_estimators=300, max_depth=5, tree_method='hist', random_state=42)
model_typhoid.fit(X_train_t, y_train_t, sample_weight=sample_weights)

acc_typhoid = accuracy_score(y_test_t, model_typhoid.predict(X_test_t))
print(f"✅ TYPHOID ACCURACY : {acc_typhoid * 100:.2f}%")

print("\n" + "="*50)
print("🎯 TROPICAL MODELS RE-EVALUATED SUCESSFULLY!")
print("="*50)
```
